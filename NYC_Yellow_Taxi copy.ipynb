Entre sus principales aplicaciones se encuentran el análisis exploratorio de datos, la identificación de patrones de movilidad, el uso de técnicas de muestreo y particionamiento, la construcción de modelos supervisados para predicción de tarifas y la aplicación de modelos no supervisados para segmentación de viajes.

# 2. Comparación de Datasets con la Justificación de la Selección

En contraste, el NYC Taxi Trip Records Dataset ofrece un mejor balance para los objetivos del proyecto. Sus datos combinan información temporal, geográfica, económica y de comportamiento, lo que permite abordar el análisis desde diferentes perspectivas. A diferencia del dataset de vuelos, ofrece mayor variedad analítica y permite explorar patrones más diversos.

# 6. Métricas recomendadas para el análisis

## Métricas descriptivas

- Número total de viajes.
- Número de viajes por mes, día y hora.
- Número de viajes por zona de origen y destino.
- Promedio, mediana, mínimo y máximo de `trip_distance`.

- Distribución por `payment_type`.
- Distribución por `RatecodeID`.

## Métricas de calidad de datos

- Porcentaje de valores nulos por columna.
- Número de registros duplicados.
- Número de viajes con distancia igual a cero.
- Número de viajes con importes negativos.
- Número de viajes con duración cero o negativa.
- Número de registros con `passenger_count` nulo.
- Número de registros con zonas de origen o destino sin correspondencia en la tabla auxiliar.
- Porcentaje de registros eliminados o filtrados durante la limpieza.

## Métricas para modelos de regresión

- **MAE**: mide el error absoluto promedio.
- **RMSE**: penaliza errores grandes.
- **R²**: mide la proporción de variación explicada por el modelo.

Estas métricas pueden usarse para modelos de predicción de `total_amount`, `fare_amount`, `tip_amount` o duración del viaje.

## Métricas para modelos de clasificación

- Accuracy.
- Precision.
- Recall.
- F1-score.
- Matriz de confusión.

Estas métricas aplican si se define un problema de clasificación, por ejemplo viaje con propina contra viaje sin propina, pago con tarjeta contra efectivo, o viaje corto contra viaje largo.

## Métricas para clustering

- Silhouette Score.
- Cantidad de registros por clúster.
- Promedio de distancia, tarifa y duración por clúster.
- Interpretabilidad de cada segmento.

Estas métricas aplican si se busca segmentar viajes con aprendizaje no supervisado.

## Dataset

```python
# @title Dataset Utils

BASE_URL = "https://d37ci6vzurychx.cloudfront.net/trip-data"

def taxi_file_url(taxi_type: str, year: int, month: int) -> str:
    """
    taxi_type:
      - yellow
      - green
      - fhv
      - fhvhv
    """
    return f"{BASE_URL}/{taxi_type}_tripdata_{year}-{month:02d}.parquet"


def download_taxi_data(
    taxi_type: str,
    years_months: list[tuple[int, int]],
    output_dir: str = "/content/data/taxi",
    overwrite: bool = False
):
    Path(output_dir).mkdir(parents=True, exist_ok=True)

    downloaded_files = []

    for year, month in years_months:
        url = taxi_file_url(taxi_type, year, month)
        filename = url.split("/")[-1]
        output_path = Path(output_dir) / filename

        if output_path.exists() and not overwrite:
            print(f"Already exists: {output_path}")
            downloaded_files.append(str(output_path))
            continue

        print(f"Downloading: {url}")

        response = requests.get(url, stream=True, timeout=120)

        if response.status_code != 200:
            print(f"Could not download {url} - status {response.status_code}")
            continue

        with open(output_path, "wb") as f:
            for chunk in response.iter_content(chunk_size=1024 * 1024):
                if chunk:
                    f.write(chunk)

        print(f"Saved: {output_path}")
        downloaded_files.append(str(output_path))

    return downloaded_files
```

```python
#@title Leer el dataset — schema-evolution-aware reader
import glob as pyglob
import pyarrow.parquet as pq
from functools import reduce

# findspark.init()

spark = (
    SparkSession.builder
    .appName("NYC Yellow Taxi")
    .getOrCreate()
)

# NYC TLC changed passenger_count from DOUBLE to INT64 around 2025-05.
# Spark allocates typed mutable slots at schema inference time and has no
# implicit DOUBLE<->INT64 coercion, so a single glob read crashes on the
# first mismatched file.
#
# Fix: use PyArrow to read only the Parquet footer metadata (fast, no data
# scan) to bucket files by physical type, then normalize to DoubleType
# before union so the final schema is consistent throughout.

all_files = sorted(pyglob.glob(f"{drive_dir}/yellow_tripdata_*.parquet"))

double_pc_files, long_pc_files = [], []
for fpath in all_files:
    pq_schema = pq.read_schema(fpath)
    idx = pq_schema.get_field_index("passenger_count")
    if idx >= 0 and str(pq_schema.field(idx).type) == "int64":
        long_pc_files.append(fpath)
    else:
        double_pc_files.append(fpath)

print(f"passenger_count DOUBLE : {len(double_pc_files)} files")
print(f"passenger_count INT64  : {len(long_pc_files)} files")

dfs = []
if double_pc_files:
    dfs.append(spark.read.parquet(*double_pc_files))
if long_pc_files:
    dfs.append(
        spark.read.parquet(*long_pc_files)
        .withColumn("passenger_count", col("passenger_count").cast(T.DoubleType()))
    )

# allowMissingColumns=True handles any extra columns added/removed between
# schema versions by filling them with nulls rather than raising an error.
df = reduce(lambda a, b: a.unionByName(b, allowMissingColumns=True), dfs)
```

```python
# Probando si no se desborda la memoria

# --- Descargar taxi_zone_lookup.csv si no existe ---
zones_csv_path = f"{drive_dir}/taxi_zone_lookup.csv"
zones_csv_url  = "https://d37ci6vzurychx.cloudfront.net/misc/taxi_zone_lookup.csv"

if not Path(zones_csv_path).exists():
    print(f"Downloading zones lookup: {zones_csv_url}")
    response = requests.get(zones_csv_url, timeout=30)
    response.raise_for_status()
    Path(zones_csv_path).write_bytes(response.content)
    print(f"Saved: {zones_csv_path}")
else:
    print(f"Already exists: {zones_csv_path}")

zones = spark.read.csv(
    zones_csv_path,
    header=True,
    inferSchema=True
)

from pyspark.sql.functions import (
    col, hour, when, count, avg
)

df_test = df.select(
    "tpep_pickup_datetime",
    "trip_distance",
    "fare_amount",
    "PULocationID"
)

# Obtenemos la hora de inicio del viaje
df_test = df_test.withColumn(
    "hour",
    hour("tpep_pickup_datetime")
)

# Transfomrmamos la distancia asumiendo
# que menos de 2 millas es corto
# entre 2 y 10 es mediano
# mas de 10 es largo
df_test = df_test.withColumn(
    "distance_segment",
    when(col("trip_distance") < 2, "short")
    .when(col("trip_distance") < 10, "medium")
    .otherwise("long")
)

# Cambiando los IDs por nombre de la zona
df_test = df_test.join(
    zones,
    df_test.PULocationID == zones.LocationID,
    "left"
)

# Agregacion, contamos los viajes y obtenemos el promedio por zona, hora
# y segmento de distancia
result = df_test.groupBy(
    "hour",
    "Borough",
    "distance_segment"
).agg(
    count("*").alias("num_trips"),
    avg("fare_amount").alias("avg_fare")
)

# Ordenar por cantidad de viajes
result = result.orderBy(col("num_trips").desc())

# Ejecutar accion
result.show(50, truncate=False)
```

```python
# ============================================================
# 2. Análisis de Agregaciones
# ============================================================

print("=" * 70)
print("2. ANÁLISIS DE AGREGACIONES Y AGRUPACIONES")
print("=" * 70)

# 2.1  Suma(total_amount, fare_amount) y Promedio(trip_distance, passenger_count) por VendorID
print()
print("--- 2.1  Suma y Promedio por VendorID ---")
(
    df.groupBy("VendorID")
    .agg(
        F.count("*").alias("num_viajes"),
        F.sum("total_amount").alias("total_amount_sum"),
        F.sum("fare_amount").alias("fare_amount_sum"),
        F.avg("trip_distance").alias("avg_distancia"),
        F.avg("passenger_count").alias("avg_pasajeros"),
    )
    .orderBy("VendorID")
    .show(truncate=False)
)

# 2.2  Sumatoria de total_amount por fecha de recogida (solo años 2023-2026)
print()
print("--- 2.2  Sumatoria de total_amount por tpep_pickup_datetime (agrupado por día, años 2023-2026) ---")
(
    df.filter(F.year("tpep_pickup_datetime").between(2023, 2026))
    .withColumn("fecha_recogida", F.to_date("tpep_pickup_datetime"))
    .groupBy("fecha_recogida")
    .agg(
        F.count("*").alias("num_viajes"),
        F.sum("total_amount").alias("total_amount_sum"),
    )
    .orderBy("fecha_recogida")
    .show(200, truncate=False)
)

# 2.3  Sumatoria de total_amount por año y mes
print()
print("--- 2.3  Sumatoria de total_amount por Año y Mes ---")
(
    df.groupBy(
        F.year("tpep_pickup_datetime").alias("año"),
        F.month("tpep_pickup_datetime").alias("mes"),
    )
    .agg(
        F.count("*").alias("num_viajes"),
        F.sum("total_amount").alias("total_amount_sum"),
    )
    .orderBy("año", "mes")
    .show(50, truncate=False)
)

# 2.4  Sumatoria de tip_amount por año y mes con total_amount y porcentaje estimado de propina
print()
print("--- 2.4  Sumatoria de tip_amount por Año y Mes (con total_amount y % propina estimado) ---")
(
    df.groupBy(
        F.year("tpep_pickup_datetime").alias("año"),
        F.month("tpep_pickup_datetime").alias("mes"),
    )
    .agg(
        F.count("*").alias("num_viajes"),
        F.sum("total_amount").alias("total_amount_sum"),
        F.sum("tip_amount").alias("tip_amount_sum"),
        F.round(
            F.sum("tip_amount") * 100.0 / F.sum("total_amount"), 2
        ).alias("tip_pct_estimado"),
    )
    .orderBy("año", "mes")
    .show(50, truncate=False)
)
```

```python
# ============================================================
# REPORTE INTEGRAL DE ANÁLISIS DEL DATASET
# ============================================================

SEP  = "=" * 70
SEP2 = "-" * 70

print(SEP)
print("REPORTE INTEGRAL DE ANÁLISIS DEL DATASET")
print(SEP)

# 1. IDENTIFICACIÓN DEL DATASET
print()
print("1. IDENTIFICACIÓN DEL DATASET")
print(SEP2)
print("  Nombre   : NYC Yellow Taxi Trip Records")
print("  Formato  : Parquet (múltiples archivos)")
print("  Fuente   : NYC Taxi & Limousine Commission (TLC)")
print("  URL      : https://www.nyc.gov/site/tlc/about/tlc-trip-record-data.page")
date_range = df.agg(
    spark_min("tpep_pickup_datetime").alias("min_date"),
    spark_max("tpep_pickup_datetime").alias("max_date"),
).collect()[0]
print(f"  Desde    : {date_range['min_date']}")
print(f"  Hasta    : {date_range['max_date']}")
print(f"  Columnas : {len(df.columns)}")

# 2. ESTADÍSTICAS GENERALES
print()
print("2. ESTADÍSTICAS GENERALES")
print(SEP2)
_num_rows    = df.count()
_size_bytes  = df._jdf.queryExecution().optimizedPlan().stats().sizeInBytes()
print(f"  Total de registros  : {_num_rows:,}")
print(f"  Total de columnas   : {len(df.columns)}")
print(f"  Tamaño estimado     : {_size_bytes / 1024**3:.2f} GB")

# 3. VALORES FALTANTES Y CALIDAD DE DATOS
print()
print("3. VALORES FALTANTES Y CALIDAD DE DATOS")
print(SEP2)
_nan_compat = {"double", "float"}
_null_counts = df.select([
    F.count(F.when(
        (F.isnan(c) if t in _nan_compat else F.lit(False)) | F.col(c).isNull(), c
    )).alias(c)
    for c, t in df.dtypes
]).toPandas().iloc[0].to_dict()
_cols_nulls = {k: int(v) for k, v in _null_counts.items() if int(v) > 0}
print(f"  Columnas con nulos/NaN  : {len(_cols_nulls)}/{len(df.columns)}")
print(f"  Total de celdas vacías  : {sum(_cols_nulls.values()):,}")
print()
for _cn, _cnt in sorted(_cols_nulls.items(), key=lambda x: -x[1]):
    _pct = _cnt / _num_rows * 100
    print(f"  {_cn:<30}: {_cnt:>12,}  ({_pct:.2f}%)")

# 4. RESUMEN DE DATOS NUMÉRICOS
print()
print("4. RESUMEN DE DATOS NUMÉRICOS")
print(SEP2)
_key_num = ["trip_distance", "fare_amount", "total_amount",
            "tip_amount", "tolls_amount", "passenger_count"]
_avail_num = [c for c in _key_num if c in set(df.columns)]
df.select(_avail_num).describe().show(truncate=False)

# 5. ANÁLISIS DE CARACTERÍSTICAS CLAVE
print()
print("5. ANÁLISIS DE CARACTERÍSTICAS CLAVE")
print(SEP2)
print("  Distribución de VendorID:")
df.groupBy("VendorID").count().orderBy("VendorID").show()
print("  Distribución de payment_type:")
df.groupBy("payment_type").count().orderBy("payment_type").show()
print("  Distribución de RatecodeID:")
df.groupBy("RatecodeID").count().orderBy("RatecodeID").show()

# 6. PROBLEMAS CONOCIDOS Y CONSIDERACIONES
print()
print("6. PROBLEMAS CONOCIDOS Y CONSIDERACIONES")
print(SEP2)
issues = [
    {
        "Problema": "Datos de Coordenadas",
        "Descripción": "Algunos registros pueden tener latitud/longitud faltante o inválida",
        "Impacto": "Usar LocationID para análisis espacial si las coordenadas no están disponibles",
    },
    {
        "Problema": "Valores Atípicos (Outliers)",
        "Descripción": "Viajes extremadamente largos o tarifas muy altas pueden ser errores de captura",
        "Impacto": "Filtrar durante el preprocesamiento (p. ej., trip_distance < 200)",
    },
    {
        "Problema": "Valores en Cero o Negativos",
        "Descripción": "Algunos valores de tarifa o distancia pueden ser cero o negativos",
        "Impacto": "Eliminar viajes inválidos antes de aplicaciones de ML",
    },
    {
        "Problema": "Cobertura Temporal",
        "Descripción": "El dataset cubre un rango de fechas específico; las predicciones requieren datos recientes",
        "Impacto": "Los patrones estacionales pueden variar; reentrenar los modelos periódicamente",
    },
    {
        "Problema": "Cambio de Formato de Versión",
        "Descripción": (
            "La TLC modificó su formato reemplazando coordenadas geográficas "
            "por IDs de zona a partir de 2016. Este dataset conserva las coordenadas originales."
        ),
        "Impacto": "Permite clustering geoespacial. No es compatible directamente con datasets de 2016 en adelante.",
    },
]
for i, issue in enumerate(issues, 1):
    print(f"  {i}. [{issue['Problema']}]")
    print(f"     Descripción : {issue['Descripción']}")
    print(f"     Impacto     : {issue['Impacto']}")
    print()

# 7. IDONEIDAD PARA APLICACIONES DE ML
print()
print("7. IDONEIDAD PARA APLICACIONES DE ML")
print(SEP2)
_ml_apps = [
    ("Predicción de Tarifa (Regresión)",          "ALTA",  "fare_amount, trip_distance, duración"),
    ("Pronóstico de Demanda (Series de Tiempo)",  "ALTA",  "datetime, PULocationID, conteos"),
    ("Predicción de Duración del Viaje",          "ALTA",  "tiempos pickup/dropoff, distancia"),
    ("Clustering de Zonas (No Supervisado)",      "MEDIA", "PULocationID, DOLocationID"),
    ("Predicción de Propina (Clasificación)",     "MEDIA", "fare_amount, payment_type, distancia"),
    ("Detección de Anomalías",                    "MEDIA", "trip_distance, outliers en fare_amount"),
]
print(f"  {'Aplicación':<45} {'Idoneidad':<10} Características Clave")
print(f"  {'-'*45} {'-'*10} {'-'*35}")
for _app, _suit, _feats in _ml_apps:
    print(f"  {_app:<45} {_suit:<10} {_feats}")
```

```python
# ============================================================
# ESCRITURA DEL DATASET UNIFICADO EN FORMATO PARQUET
# Particionado por year/month para optimizar lecturas futuras
# ============================================================

output_path = drive_dir + "/yellow_taxi_unified"
# '/content/drive/MyDrive/Colab Notebooks/BigData/raw/taxi/yellow'
Path(output_path).mkdir(parents=True, exist_ok=True)

print("=" * 70)
print("WRITING UNIFIED DATASET TO PARQUET")
print("=" * 70)
print(f"  Output path  : {output_path}")
print(f"  Records      : {df.count():,}")
print(f"  Partitioning : year / month")
print()

(
    df
    .withColumn("year",  F.year("tpep_pickup_datetime"))
    .withColumn("month", F.month("tpep_pickup_datetime"))
    .write
    .mode("overwrite")
    .partitionBy("year", "month")
    .parquet(output_path)
)

print(f"Dataset written successfully to: {output_path}")
```
